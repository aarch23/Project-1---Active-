```python
# Importing relevant python pacakges

import pandas as pd
import numpy as np
import os
import json
from defillama import DefiLlama
```

```python
# Endpoints are accessible without an API key

# initialize api client|
llama = DefiLlama()

# Get all protocols data
response = llama.get_all_protocols()

# Get a protocol data
response_1 = llama.get_protocol(name='uniswap')

# Get historical values of total TVL
response_2 = llama.get_historical_tvl()

# Get protocol TVL
response_3 = llama.get_protocol_tvl(name='Curve')
```
